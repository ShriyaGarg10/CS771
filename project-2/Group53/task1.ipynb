```python
import os
import numpy as np
import tensorflow as tf
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from scipy.spatial.distance import mahalanobis
from scipy.linalg import inv
import torch
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files


def load_data(file_path):
    """Loads data and labels from a specified file."""
    data = torch.load(file_path, weights_only=True)
    inputs, labels = data['data'], data.get('targets')
    return inputs, labels


def save_transformed_data(features, labels, save_path):
    """Saves transformed features and labels to a file."""
    torch.save({'features': features, 'labels': labels}, save_path)


def load_transformed_data(file_path):
    """Loads preprocessed features and labels from a file."""
    data = torch.load(file_path)
    return data['features'], data.get('labels')

def plot_heatmap(data, x_labels, y_labels, title="Accuracy Heatmap", cmap="viridis",  save_path=None):
    """Generates a heap map plot"""
    plt.figure(figsize=(10, 8))
    sns.heatmap(data, annot=True, fmt=".3f", cmap=cmap, xticklabels=x_labels, yticklabels=y_labels)
    plt.title(title, fontsize=16)
    plt.xlabel("Evaluation Datasets", fontsize=14)
    plt.ylabel("Training Iterations", fontsize=14)
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')  # Saves the figure with high resolution
        print(f"Heatmap saved to {save_path}")
    plt.show()


class ImageFeatureExtractor:
    def __init__(self, target_size=(224, 224, 3)):
        # Initialize a ResNet50 model for feature extraction
        self.base_model = tf.keras.applications.ResNet50(
            weights='imagenet', include_top=False, pooling='avg', input_shape=target_size
        )

    def preprocess_images(self, images):
        """
        Adjust images to the format expected by ResNet50.
        """
        images = images.astype(np.float32)
        resized_images = np.array([tf.image.resize(img, (224, 224)).numpy() for img in images])
        return tf.keras.applications.resnet50.preprocess_input(resized_images)

    def extract_features(self, images):
        """
        Generate deep features from the provided images.
        """
        preprocessed_images=self.preprocess_images(images)
        return self.base_model.predict(preprocessed_images, batch_size=32, verbose=1)


class DimensionReducer:
    def __init__(self, components=256):
        self.reducer=PCA(n_components=components)

    def fit_transform(self, features):
        return self.reducer.fit_transform(features)

    def transform(self, features):
        return self.reducer.transform(features)


class PrototypeClassifier:
    def __init__(self, num_categories):
        self.num_categories = num_categories
        self.centroids=None
        self.inv_covariance_matrix=None

    def train(self, features, labels):
        """Compute class prototypes and covariance matrix."""
        self.centroids = []
        class_covariances = []

        for category in range(self.num_categories):
            class_data = features[labels == category]
            class_centroid = class_data.mean(axis=0)
            self.centroids.append(class_centroid)

            if len(class_data) > 1:  # Ensure sufficient data for covariance
                covariance_matrix=np.cov(class_data,rowvar=False)
                class_covariances.append(covariance_matrix)

        pooled_covariance=np.mean(class_covariances, axis=0)
        self.inv_covariance_matrix=inv(pooled_covariance)
        self.centroids = np.array(self.centroids)

    def classify(self, features):
        """Assign class labels based on Mahalanobis distance."""
        distances = np.array([
            [mahalanobis(x, centroid, self.inv_covariance_matrix) for centroid in self.centroids]
            for x in features
        ])
        return np.argmin(distances, axis=1)

    def refine(self, features, predicted_labels):
        """Refine centroids using predicted labels."""
        for category in range(self.num_categories):
            category_data = features[predicted_labels == category]
            if len(category_data) > 0:
                category_mean = category_data.mean(axis=0)
                self.centroids[category] = category_mean


def preprocess_and_store_data(file_paths, extractor, output_dir):
    """
    Precompute and save features for datasets.
    """
    os.makedirs(output_dir, exist_ok=True)
    for file_path in file_paths:
        file_name = os.path.basename(file_path).replace('.tar.pth', '_processed.pkl')
        save_location = os.path.join(output_dir, file_name)

        if os.path.exists(save_location):
            print(f"Features for {file_path} already exist. Skipping extraction.")
            continue

        raw_data, raw_labels = load_data(file_path)
        extracted_features = extractor.extract_features(raw_data)
        save_transformed_data(extracted_features, raw_labels, save_location)
        print(f"Processed features saved to {save_location}")


def iterative_training(train_paths, eval_paths, feature_folder, class_count=10):
    score_matrix = np.zeros((len(train_paths), len(eval_paths)))
    reducer = DimensionReducer()
    classifier = PrototypeClassifier(num_categories=class_count)

    initial_features, initial_labels = load_transformed_data(
        os.path.join(feature_folder, '1_train_data_processed.pkl')
    )
    reduced_initial_features = reducer.fit_transform(initial_features)
    classifier.train(reduced_initial_features, np.array(initial_labels))

    for stage in range(1, len(train_paths) + 1):
        print(f"Training iteration {stage}...")

        for eval_index in range(stage):
            eval_features, eval_labels = load_transformed_data(
                os.path.join(feature_folder, f'{eval_index + 1}_eval_data_processed.pkl')
            )
            reduced_eval_features = reducer.transform(eval_features)
            predictions = classifier.classify(reduced_eval_features)
            score_matrix[stage - 1, eval_index] = accuracy_score(eval_labels, predictions)

        if stage == len(train_paths):
            break

        next_train_features, _ = load_transformed_data(
            os.path.join(feature_folder, f'{stage + 1}_train_data_processed.pkl')
        )
        reduced_next_train_features = reducer.transform(next_train_features)
        predicted_labels = classifier.classify(reduced_next_train_features)

        classifier.refine(reduced_next_train_features, predicted_labels)

    return score_matrix, reducer, classifier


# Define the upload directory where files are placed
upload_dir = "/content/"

# Update file paths to point to the Colab uploads folder
training_data_paths = [os.path.join(upload_dir, f"{i}_train_data.tar.pth") for i in range(1, 11)]
evaluation_data_paths = [os.path.join(upload_dir, f"{i}_eval_data.tar.pth") for i in range(1, 11)]

# Define the directory to save processed features
processed_feature_dir = os.path.join(upload_dir, "processed_data")
os.makedirs(processed_feature_dir, exist_ok=True)  # Ensure directory exists


# Proceed with feature extraction and training
feature_extractor = ImageFeatureExtractor()
preprocess_and_store_data(training_data_paths, feature_extractor, processed_feature_dir)
preprocess_and_store_data(evaluation_data_paths, feature_extractor, processed_feature_dir)

accuracy_results, dim_reducer, model_classifier = iterative_training(
    training_data_paths, evaluation_data_paths, processed_feature_dir
)

x_labels = [f"Eval {i+1}" for i in range(accuracy_results.shape[1])]
y_labels = [f"Train {i+1}" for i in range(accuracy_results.shape[0])]
save_path= os.path.join(upload_dir, "task1.png")
plot_heatmap(accuracy_results, x_labels, y_labels, save_path=save_path)
files.download(save_path)
```
